# 7. Matplotlib ve Seaborn

## 7.1 Import Matplotlib, Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 7.2 Verinin Okunması

```python
data = pd.read_csv("iris.csv")
data.head()
```

## 7.3 Görselleştirme Teknikleri

### Scatterplot

```python
sns.scatterplot(data=data, x="Id", y="SepalWidthCm", hue="Species")
```

### Lineplot

```python
sns.lineplot(data=data, x="Id", y="SepalWidthCm", hue="Species")
```

### Barplot

```python
data.head()
```

```python
data.groupby("Species").mean()
```

```python
data.groupby("Species").mean()["SepalWidthCm"]
```

```python
sns.barplot(data=data, x="Species", y="SepalWidthCm")
```

### 7.6 Pie Chart

```python
data_grouped = data.groupby("Species").mean()["PetalWidthCm"]
data_grouped
```

```python
data_grouped.plot.pie()
```

```python
plt.figure(figsize=(10,10))
data_grouped.plot.pie()
```
